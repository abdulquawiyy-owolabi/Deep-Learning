```python
def operation(a1: torch.Tensor, a2: torch.Tensor, a3: torch.Tensor):
  answer = a1@a2+a3
  return answer
```

```python
a1 = torch.tensor([[2, 4], [5, 7]])
a2 = torch.tensor([[1, 1], [2, 3]])
a3 = torch.tensor([[10, 10], [12, 1]])

a = operation(a1, a2, a3)
print(a)
```

```python
def dot_product(b1: torch.Tensor, b2: torch.Tensor):
  answer = torch.dot(b1, b2)
  return answer
```

```python
b1 = torch.tensor([3, 5, 7])
b2 = torch.tensor([2, 4, 8])
## Uncomment to test your function
b = dot_product(b1, b2)
print(b)
```
